# 주소 -> 좌표 변환

```python
df = pd.read_csv("서울 소방서 현황(2017.12.31).csv", engine='python')
df
```

```python
df['위치정보(X)'] = None
df['위치정보(Y)'] = None
df.head()
```

```python
from collections import namedtuple
import requests
from bs4 import BeautifulSoup
import urllib.request
from urllib import parse
import json
from lxml import etree

url = "https://maps.googleapis.com/maps/api/geocode/json"
params = {
    "address" : None,
    "key" : "AIzaSyBW3kWikDl63OaOC99TXQS3r1XQ4szW_2g",
}

def getData(url, params):
    html = requests.get(url, params)
    print(html.text)
    data= BeautifulSoup(html.text, "lxml")
    
    return data

for i in range(0, len(df)):
    params["address"] = df["주소"][i]
    data = getData(url, params)
    newDictionary=json.loads(str(data.text))
    try:
        lat = newDictionary["results"][0]['geometry']['location']['lat']
        lng =  newDictionary["results"][0]['geometry']['location']['lng']
        df['위치정보(X)'].loc[i] = lat
        df['위치정보(Y)'].loc[i] = lng
    except:
        print(i + "에러")
```

```python
df.to_csv("서울 소방서 현황(2017.12.31).csv", encoding='cp949', index=False)
```
