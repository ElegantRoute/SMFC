```python
import re
import time
```

```python
from selenium import webdriver
```

```python
p1 = re.compile(r'서울특별시 [가-힝]+구 +[가-힝]+.+')
p2 = re.compile(r'서울특별시 [가-힣]+구 [가-힣0-9]+[동|가]')
pat = re.compile(r'(서울[가-힝]+ +[가-힝]+ *[0-9가-힝-,]+ *?\d* ?\d*).*')
```

```python
driver = webdriver.Chrome('C:/Users/kyujo/python/chromedriver_win32/chromedriver.exe')
driver.get('http://www.juso.go.kr/support/AddressMainSearch.do?searchType=TOTAL')
```

```python
dong=[]
for j in data['주소']:
#     print(j)
    j = pat.sub(r'\g<1>',j)
    time.sleep(0.3)
    if p1.match(j) != None:#도로명 주소이면
        driver.find_element_by_id('keyword').click()
        driver.find_element_by_id('keyword').send_keys(j)
        driver.find_element_by_id('searchButton').click()
        try:#지번주소 긁어옴
            body=driver.find_element_by_css_selector('div.num.landLot')
            dong.append(p2.match(body.text).group())
            driver.find_element_by_id('keyword').clear()
        except:#지번주소가 없으면 그냥 넣는다# 얘는 else로 안감
            dong.append(j)
            driver.find_element_by_id('keyword').clear()
    else:#도로명 주소가 아니면(지번주소면) 그냥 넣음
        dong.append(j)
dong
```

```python
data['동주소']=dong
```

```python
data.drop(columns='동주소',inplace=True)
```

```python
driver = webdriver.Chrome('C:/Users/kyujo/python/chromedriver_win32/chromedriver.exe')
driver.get('http://map.daum.net/')
```

```python
p1 = re.compile(r'(서울)[가-힣]+ ?(.+)')
p2 = re.compile(r'(.+) +?지번')
# pat = re.compile(r'(서울[가-힝]+ +[가-힝]+ *[0-9가-힝-,]+ *?\d* ?\d*).*')
```

```python
# 수정

add=[]
driver.find_element_by_id('search.keyword.query').clear()
for i,j in enumerate(data['경찰서명']):
    driver.find_element_by_id('search.keyword.query').click()
    driver.find_element_by_id('search.keyword.query').send_keys(j)
    driver.find_element_by_id('search.keyword.submit').click()
    time.sleep(0.2)
    try:
        body=driver.find_element_by_css_selector('p.newAddress') 
    except:
        driver.find_element_by_id('search.keyword.query').clear()
        time.sleep(0.2)
        j = p1.sub(r'\g<1>\g<2>',j)
        driver.find_element_by_id('search.keyword.query').send_keys(j)
        driver.find_element_by_id('search.keyword.submit').click()
        time.sleep(0.2)
        body=driver.find_element_by_css_selector('p.newAddress')
        if body.text != '' :
            k = p2.sub(r'\g<1>',body.text)
            add += [k]
            driver.find_element_by_id('search.keyword.query').clear()
            time.sleep(0.5)
        else:
            add += [data.iloc[i,4]]
    else:
        if body.text != '' :
            k = p2.sub(r'\g<1>',body.text)
            add += [k]
            driver.find_element_by_id('search.keyword.query').clear()
            time.sleep(0.5)
        else:
            add += [data.iloc[i,4]]
add
```

```python
data['좌표주소']=add
```
