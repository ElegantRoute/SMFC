```python
import pandas as pd
```

```python
data = pd.read_csv("2016 구별 범죄통계.csv", encoding='cp949', engine='python')
data.head()
```

```python
len(data)
```

```python
data_sorted = data.sort_values(by=['시군구'])
data_sorted.rename(columns = {"시군구":"법정구"}, inplace=True)
data_sorted
```

```python
df = pd.read_excel("statistics.xlsx", engint='python')
df.head()
```

```python
df["법정구"] = None
```

```python
df.head()
```

```python
split_addr = df["법정동"][0].split(" ")
split_addr[1] 
```

```python
for i in range(0, len(df)):
    split_addr = df["법정동"][i].split(" ")
    gu = split_addr[1]
    df["법정구"][i] = gu
```

```python
df
```

```python
dfgu = df.groupby("법정구").sum()
```

```python
len(dfgu)
```

```python
dfgu_sorted = dfgu.sort_values(by=['법정구'])
dfgu_sorted
```

```python
merge_df = pd.merge(dfgu_sorted, data_sorted, on='법정구')
```

```python
merge_df.to_excel("statistics_merge.xlsx", encoding='cp949', index=False)
```
