```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
from selenium import webdriver
import time

driver = webdriver.Chrome()
```

```python
# 데이터 불러오기
df = pd.read_csv("법정동끝/서울아동안전지킴이집.csv", engine='python', encoding='cp949')
```

```python
df["포맷주소"] = None
df["법정동 주소"] = None
df["법정동"] = None
df.head()
```

```python
for i in range(0, len(df)):
    addr = df["주소"][i]
    name = df["이름"][i]
    #addr = name + " " + addr
    #print(addr)
    driver.implicitly_wait(30)
    url = "https://maps.googleapis.com/maps/api/geocode/xml?address="
    key = "&key=AIzaSyBW3kWikDl63OaOC99TXQS3r1XQ4szW_2g"
    url = url + addr + key
    #print(url)
    driver.implicitly_wait(30)
    driver.get(url)
    data = driver.find_elements_by_css_selector("#collapsible1 div.expanded div.collapsible-content")
    line = data[0].text
    line = line.split("<formatted_address>")[1]
    line = line.split("</formatted_address>")[0]
    #print(line)
    #print(data)
    #driver.implicitly_wait(30)
    df["포맷주소"][i] = line
    formataddr = line
    formataddr2 = formataddr.split()
    #print(formataddr)
    try:
        df["법정동 주소"][i] = formataddr2[1] + " " + formataddr2[2] + " " + formataddr2[3]
        df["법정동"][i] = formataddr2[3]
    except:
        df["법정동 주소"][i] = 0
        df["법정동"][i] = 0
```

```python
x = df["x좌표"][0]
y = df["y좌표"][0]
latlng = str(x) + "," + str(y)
latlng
```

```python
# 좌표로 주소 뽑기
for i in range(0, len(df)):
    x = df["x좌표"][i]
    y = df["y좌표"][i]
    latlng = str(x) + "," + str(y)

    #addr = name + " " + addr
    #print(addr)
    driver.implicitly_wait(30)
    url = "https://maps.googleapis.com/maps/api/geocode/xml?latlng="
    key = "&key=AIzaSyBW3kWikDl63OaOC99TXQS3r1XQ4szW_2g"
    url = url + latlng + key
    print(url)
    driver.implicitly_wait(30)
    driver.get(url)
    data = driver.find_elements_by_css_selector("#collapsible1 div.expanded div.collapsible-content")
    line = data[0].text
    line = line.split("<formatted_address>")[1]
    line = line.split("</formatted_address>")[0]
    #print(line)
    #print(data)
    #driver.implicitly_wait(30)
    df["포맷주소"][i] = line
    formataddr = line
    formataddr2 = formataddr.split()
    #print(formataddr)
    try:
        df["법정동 주소"][i] = formataddr2[1] + " " + formataddr2[2] + " " + formataddr2[3]
        df["법정동"][i] = formataddr2[3]
    except:
        df["법정동 주소"][i] = 0
        df["법정동"][i] = 0
```
