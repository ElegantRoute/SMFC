```python
import pandas as pd
```

```python
data = pd.read_excel('project/2018 2분기 동, 연령별 인구현황.xls')
```

```python
data = data[['기간','자치구','행정동','구분','계']]
```

```python
data
```

```python
data = data[data['행정동']!='소계']
```

```python
data = data[data['행정동']!='합계']
```

```python
data = data[data['구분']!= '계']
```

```python
len(data['행정동'].unique())
```

```python
data.groupby(['자치구','행정동'])['계'].sum()
```

```python
data.to_csv('dong_population.csv',encoding='cp949')
```
