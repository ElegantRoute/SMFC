```python
import pandas as pd

df = pd.read_csv("서울아동보호치료시설.csv", engine='python', encoding='utf-8')
#df.drop("Unnamed: 0", axis=1, inplace=True)
df.head()
df.tail(20)
```

```python
from collections import namedtuple
import requests
from bs4 import BeautifulSoup
import urllib.request
from urllib import parse
import json
from lxml import etree

# https://maps.googleapis.com/maps/api/geocode/json
# ?address=1600+Amphitheatre+Parkway,+Mountain+View,+CA
# &key=AIzaSyBW3kWikDl63OaOC99TXQS3r1XQ4szW_2g



def getData(url):
    html = requests.get(url)
    print(html.text)
    data= BeautifulSoup(html.text, "lxml")
    
    return data
```

```python
url = "https://maps.googleapis.com/maps/api/geocode/json?address="
key = 'AIzaSyBW3kWikDl63OaOC99TXQS3r1XQ4szW_2g'
```

```python
# 구글맵이 서울특별시 중랑구 면목동,412-3을
# 주소 + 이름으로 검색했을 때, 주소만가지고 검색했을 때랑 다르게 이름에 우선순위를 두는건지, 
# 주소에 해당하는 이름을 갖은 건물이 없어서인지 
# 이상한 주소를 검색해옴.;;;
# 결국, csv파일의 좌표값이 이상한 좌표값을 갖는 애만, slicing으로 찾아서 구글맵에서 가져온 좌표로 수정하기로 함.
for i in range(0, 1):
    name = df["이름"][i]
    addr = df["주소"][i]
    url = "https://maps.googleapis.com/maps/api/geocode/json?address="
    key = 'AIzaSyBW3kWikDl63OaOC99TXQS3r1XQ4szW_2g'
    url = url + name + addr + "&key=" + key
    print(url)
    data = getData(url)
    newDictionary=json.loads(str(data.text))
    try:
        lat = newDictionary["results"][0]['geometry']['location']['lat']
        lng =  newDictionary["results"][0]['geometry']['location']['lng']
        df['위치정보(X)'].loc[i] = lat
        df['위치정보(Y)'].loc[i] = lng
        #print(lat)
        #pirnt(lng)
    except:
        df['위치정보(X)'].loc[i] = df['x좌표'].loc[i]
        df['위치정보(Y)'].loc[i] = df['y좌표'].loc[i]
```
